```python
X = df.drop(['status','name'],axis = 1)
y=df.status
y
```

```python
pip install imblearn
```

```python
#detect the label imbalance
from imblearn.over_sampling import RandomOverSampler #better. not going to loose important data
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

print(Counter(y))
```

```python
#balance the labels
ros = RandomOverSampler()
x_ros, y_ros = ros.fit_resample(X,y)
print(Counter(y_ros))
```

```python
#scaling is important in the algorithms such as suppot vector machine and k-nearest neighbours where distance between data points are important
scaler = MinMaxScaler((-1,1))
x = scaler.fit_transform(x_ros)
y = y_ros
```

```python
#feature engineering
from sklearn.decomposition import PCA
pca = PCA(.95)
X_PCA = pca.fit_transform(x)
print(x.shape)
print(X_PCA.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_PCA, y, test_size = 0.2, random_state = 7)
```

```python
#linear regression

from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(C=0.4, max_iter = 1000, solver = 'liblinear')
lr = classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
accuracy_lr = accuracy_score(y_test, y_pred)
accuracy_lr
```

```python
#decision tree

from sklearn.tree import DecisionTreeClassifier
classifier2 = DecisionTreeClassifier(random_state=14)
dt = classifier2.fit(x_train, y_train)
y_pred2 = classifier2.predict(x_test)
accuracy_dt = accuracy_score(y_test, y_pred2)
accuracy_dt
```

```python
#random forest - information gain

from sklearn.ensemble import RandomForestClassifier
classifier3 = RandomForestClassifier(random_state=14)
rf = classifier3.fit(x_train, y_train)
y_pred3 = classifier3.predict(x_test)
accuracy_rf = accuracy_score(y_test, y_pred3)
accuracy_rf
```

```python
#random forest - entropy

from sklearn.ensemble import RandomForestClassifier
classifier4 = RandomForestClassifier(criterion='entropy')
rfe = classifier4.fit(x_train, y_train)
y_pred4 = classifier4.predict(x_test)
accuracy_rfe = accuracy_score(y_test, y_pred4)
accuracy_rfe
```

```python
#svm

from sklearn.svm import SVC
model_svm = SVC(cache_size=100)
svm = model_svm.fit(x_train, y_train)
y_pred5 = model_svm.predict(x_test)
accuracy_svc = accuracy_score(y_test, y_pred5)
accuracy_svc
```

```python
#knn

from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=3)
knn = model_knn.fit(x_train, y_train)
y_pred6 = model_knn.predict(x_test)
accuracy_knn = accuracy_score(y_test, y_pred6)
accuracy_knn
```

```python
#gaussion naive bayes

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb = gnb.fit(x_train, y_train)
y_pred7 = gnb.predict(x_test)
accuracy_gnb = accuracy_score(y_test, y_pred7)
accuracy_gnb
```

```python
#bernoulli naive bayes

from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb = bnb.fit(x_train, y_train)
y_pred8 = bnb.predict(x_test)
accuracy_bnb = accuracy_score(y_test, y_pred8)
accuracy_bnb
```

```python
#XGBClassifier


model_xgb = XGBClassifier()
model_xgb.fit(x_train, y_train)
y_pred9 = model_xgb.predict(x_test)
accuracy_xgb = accuracy_score(y_test, y_pred9)
accuracy_xgb
```
